# Import relevant python packages

```python
import numpy as np #Numerical package for arrays, algebra...
%pylab notebook 
from mpl_toolkits.mplot3d import Axes3D #generate 3D plots
from scipy.special import j0,y0, j1, y1 #0th and 1st order Bessel function of first and second kind
import time
```

# $G_0(r-r^{'},\omega)=\frac{|\omega|}{4\hbar^2 {v_f}^2}[f_0(\omega,\rho)\sigma_0 + f_1(\omega,\rho) \vec{\sigma}\cdot\vec{\rho}]$ <br> 
$f_0(\omega,\rho)=sgn(\omega) Y_0(\zeta)-iJ_0(\zeta)\Theta(1-\frac{|\omega|}{\Lambda})$
<br>
$f_1(\omega,\rho)=iY_1(\zeta)+sgn(\omega) J_1(\zeta)\Theta(1-\frac{|\omega|}{\Lambda})$
<br>
$\vec{\rho}=\vec{r}-\vec{r'}~~;~~\zeta=\frac{|\omega|\rho}{\hbar v_f}$

\begin{equation*}
\sigma_x =\begin{pmatrix}
0 & 1 \\ 1 & 0
\end{pmatrix} ~~;~~ 
\sigma_y =\begin{pmatrix}
0 & -i \\ i & 0
\end{pmatrix}
~~;~~ 
\sigma_z =\begin{pmatrix}
1 & 0 \\ 0 & -1
\end{pmatrix}
\end{equation*}


```python
npix=75
norm=1.
eps=1e-8
def return_G0(x_tip,y_tip,k=1):
    G0=np.zeros((npix,npix),dtype=np.complex)
    x,y=np.indices(G0.shape)
    r=eps+np.sqrt((x-x_tip)**2.+(y-y_tip)**2.)
    G0=-norm*(y0(k*r)+1j*j0(k*r))
    return G0

G0all=np.zeros((npix**2,npix**2),dtype=np.complex)
s=0
for itip in range(npix):
    for jtip in range(npix):
        G0all[:,s]=(return_G0(x_tip=itip,y_tip=jtip,k=1.)).ravel()
        s=s+1
```

```python
x1=linspace(1,10,20)
figure()
plot(x1,heaviside(x1-5, 5))
```

```python
figure()
imshow(-1*imag(return_G0(x_tip=15,y_tip=15,k=1.)),origin="lower")
colorbar()
figure()
imshow(-imag(G0all),origin="lower")
colorbar()
```

# $\langle r |V| r^{'} \rangle$  ellipse 

```python
V=np.zeros((npix,npix),dtype=np.complex)
for i in range(npix):
    for j in range(npix):
        radius=(i-npix/2)**2/2 + (j-npix/2)**2/4
        if radius >= 70 and radius <=80:
            V[i,j]=1.

figure()
imshow(real(V),origin="lower")
colorbar()
        
Vall=diag(V.ravel())
figure()
imshow(real(Vall),origin="lower")
colorbar()
```

# $G=(\mathbb{1} -G_0V)^{-1}G_0$

```python
I0=identity(npix**2,dtype=np.complex)
# G0V=np.matmul(G0all,Vall)
G0V=np.einsum("ij,jj->ij",G0all,Vall)
Gnew=np.linalg.solve(I0-G0V,G0all)
```

# LDOS $\rho(r)=\frac{-1}{\pi}Im G(r,r)$

```python
dos=-1*imag(diag(Gnew)).reshape(npix,npix)

figure()
imshow(dos,origin="lower")
colorbar()
```

# V arc

```python
V=np.zeros((npix,npix),dtype=np.complex)
for i in range(npix/2):
    for j in range(npix/2):
        radius=(i-npix/2)**2 + (j-npix/2)**2
        if radius >= 60 and radius <=80:
            V[i,j]=1.

figure()
imshow(real(V),origin="lower")
colorbar()
```

```python
Vall=diag(V.ravel())
I0=identity(npix**2,dtype=np.complex)
# G0V=np.matmul(G0all,Vall)
G0V=np.einsum("ij,jj->ij",G0all,Vall)
Gnew=np.linalg.solve(I0-G0V,G0all)
dos=-1*imag(diag(Gnew)).reshape(npix,npix)
```

```python
figure()
imshow(dos,origin="lower")
colorbar()
```
