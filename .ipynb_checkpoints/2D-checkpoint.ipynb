```python
dos1=object_dos.dos_calculator(xnpix=40)
```

```python
dos1.return_lambda()
dos1.dbglambda
dos1.return_G0all()
```

```python
def return_gaussianV(xnpix=dos1.xnpix,ynpix=dos1.ynpix,pixwid=dos1.pixwid,x_imp=15,y_imp=15,sigma=2.,amp=0.2):
    gaussianV=np.zeros((xnpix,ynpix),dtype=np.float32)
    x,y=np.indices(gaussianV.shape)
    gaussianV=amp*(1./(2.*np.pi*sigma**2.))*np.exp(-((x-x_imp)**2.+(y-y_imp)**2.)/(2.*sigma**2.))
    return gaussianV
```

```python
V=return_gaussianV()
figure()
imshow(V,origin='lower')
colorbar()
```

```python
dos1.get_Gnew(V=V)
```
